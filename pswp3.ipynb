```python
import matplotlib.pyplot as plt
import random
%matplotlib

sampleSize = 10000
hatSize = 2027

bins = [0 for _ in range(hatSize)]
for i in range(sampleSize):
    hat = [x for x in range(1, hatSize+1)]
    hatLen = hatSize
    while hatLen > 1:
        paper1 = hat.pop(random.randrange(0, hatLen))
        paper2 = hat.pop(random.randrange(0, hatLen-1))
        paper3 = abs(paper1 - paper2)
        hat.append(paper3)
        hatLen -= 1
    bins[hat[0]] += 1

fig, ax = plt.subplots()
col = [x for x in range(0,hatSize)]
ax.bar(col, bins)
plt.show()
```
